```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#Target search for coronavirus
target = new_client.target
target_query = target.search('pde5')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#Selecttarget
selected_target = targets.target_chembl_id[2]
selected_target
```

```python
#filtering everything but entries with IC50 data
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
#Creating data frame from filtered data
df = pd.DataFrame.from_dict(res)
df
```

```python
# filtering out entries without a standard value or canonical smiles data
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
# dropping duplicates
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3.dropna
df3.to_csv('pde5_unprocessed_data.csv')
df3
```

```python
df4 = pd.read_csv('pde5_unprocessed_data.csv')
```

```python
import numpy as np
def pIC50(input):
    pIC50 = []

    for i in input['standard_value']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value', 1)
        
    return x
```

```python
df5 = pIC50(df4)
Y = df5['pIC50']
Y
```

```python
from sklearn.feature_selection import VarianceThreshold

def remove_low_variance(input_data, threshold=0.1):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]

X = remove_low_variance(X, threshold=0.1)
X
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=500, random_state=42)
model.fit(X, Y)
```
